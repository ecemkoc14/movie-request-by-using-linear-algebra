```python
import pandas as pd
import numpy as np

m_url = "https://raw.githubusercontent.com/ecemkoc14/movie-request-by-using-linear-algebra/main/en_yeni_lineer_movie.csv"
r_url = "https://raw.githubusercontent.com/ecemkoc14/movie-request-by-using-linear-algebra/main/ratings_linear(in).csv"

try:
    movies_encoded = pd.read_csv(m_url, sep=None, engine='python', encoding='latin1')
    ratings = pd.read_csv(r_url, sep=None, engine='python', encoding='latin1')

    movies_encoded.columns = movies_encoded.columns.str.strip()
    ratings.columns = ratings.columns.str.strip()

    movies_encoded.rename(columns={movies_encoded.columns[0]: 'movieId', movies_encoded.columns[1]: 'title'}, inplace=True)
    ratings.rename(columns={ratings.columns[0]: 'userId', ratings.columns[1]: 'movieId', ratings.columns[2]: 'rating'}, inplace=True)

    genres = [c for c in movies_encoded.columns if c not in ['movieId', 'title']]
    for col in genres:
        movies_encoded[col] = pd.to_numeric(movies_encoded[col], errors='coerce').fillna(0).astype(float)

    ratings['rating'] = pd.to_numeric(ratings['rating'], errors='coerce').fillna(0).astype(float)
    ratings['userId'] = pd.to_numeric(ratings['userId'], errors='coerce').fillna(0).astype(int)
    ratings['movieId'] = pd.to_numeric(ratings['movieId'], errors='coerce').fillna(0).astype(int)
    movies_encoded['movieId'] = movies_encoded['movieId'].astype(int)

    df = pd.merge(ratings, movies_encoded, on='movieId')

    for genre in genres:
        df[genre] = df[genre] * df['rating']

    user_profiles = df.groupby('userId')[genres].sum()
    user_profiles_norm = user_profiles.div(user_profiles.sum(axis=1).replace(0, 1), axis=0)

except Exception as e:
    print(f"Error: {e}")

def get_recommendations():
    try:
        val = input("User ID: ").strip()
        if not val: return
        u_id = int(float(val))

        if u_id not in user_profiles_norm.index:
            return

        user_vec = user_profiles_norm.loc[u_id].values
        movie_matrix = movies_encoded[genres].values
        scores = np.dot(movie_matrix, user_vec)

        results = movies_encoded[['movieId', 'title']].copy()
        results['score'] = scores

        watched = ratings[ratings['userId'] == u_id]['movieId'].unique()
        final_recs = results[~results['movieId'].isin(watched)].sort_values(by='score', ascending=False).head(5)

        print(final_recs[['title', 'score']])
    except:
        pass

get_recommendations()
```
