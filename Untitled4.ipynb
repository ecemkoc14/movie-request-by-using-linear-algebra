```python
import pandas as pd
import numpy as np

# 1. VERİLERİ OKUMA VE TEMİZLEME
try:
    movies_encoded = pd.read_csv('en_yeni_lineer_movie.csv', sep=None, engine='python', encoding='latin1')
    ratings = pd.read_csv('ratings_linear.csv', sep=None, engine='python', encoding='latin1')

    movies_encoded.rename(columns={movies_encoded.columns[0]: 'movieId', movies_encoded.columns[1]: 'title'}, inplace=True)
    ratings.rename(columns={ratings.columns[1]: 'movieId', ratings.columns[0]: 'userId', ratings.columns[2]: 'rating'}, inplace=True)

    genres = movies_encoded.columns[2:].tolist()

    for col in genres:
        movies_encoded[col] = pd.to_numeric(movies_encoded[col], errors='coerce').fillna(0)

    ratings['rating'] = pd.to_numeric(ratings['rating'], errors='coerce').fillna(0)
    print("Sistem başarıyla yüklendi.")
except Exception as e:
    print(f"Hata: {e}")

# 2. MATRIX ISLEMLERI (User Profile Vector Generation)
df = pd.merge(ratings, movies_encoded, on='movieId')
for genre in genres:
    df[genre] = df[genre] * df['rating']

user_profiles = df.groupby('userId')[genres].sum()
row_sums = user_profiles.sum(axis=1)
user_profiles_norm = user_profiles.div(row_sums.replace(0, 1), axis=0)

# --- ÖNERİ VE MATRİS GÖRÜNTÜLEME FONKSİYONU ---

def show_recommendation_logic():
    print("-" * 50)
    try:
        u_id = int(input("Öneri almak istediğiniz User ID'yi girin (1-100): "))
    except ValueError:
        print("Lütfen geçerli bir sayı girin!")
        return

    if u_id not in user_profiles_norm.index:
        print(f"Hata: Veri setinde {u_id} numaralı kullanıcı bulunamadı!")
        return

    # A. Kullanıcı Vektörünü Göster (User Profile Vector - U)
    print(f"\n--- [ADIM 1] Kullanıcı {u_id} için Tercih Vektörü (U) ---")
    user_vec = user_profiles_norm.loc[u_id]
    print(user_vec.to_frame().T) # Yatay düzlemde vektörü gösterir

    # B. Film Matrisini Göster (Movie Matrix - M)
    print(f"\n--- [ADIM 2] Örnek Film Matrisi (M) (İlk 5 Film) ---")
    movie_matrix = movies_encoded[genres].values
    print(pd.DataFrame(movie_matrix[:5], columns=genres, index=movies_encoded['title'][:5]))

    # C. Nokta Çarpımı İşlemi (Dot Product: R = M . U)
    print(f"\n--- [ADIM 3] Hesaplama: R = M . U (Skor = Film Vektörü * Kullanıcı Vektörü) ---")
    scores = np.dot(movie_matrix, user_vec.values)

    # Sonuçları toparla
    temp_df = movies_encoded[['movieId', 'title']].copy()
    temp_df['match_score'] = scores

    # İzlenenleri ele
    watched = ratings[ratings['userId'] == u_id]['movieId'].unique()
    recommendations = temp_df[~temp_df['movieId'].isin(watched)].sort_values(by='match_score', ascending=False).head(5)

    print("\n--- [SONUÇ] En Yüksek Skora Sahip Öneriler ---")
    print(recommendations[['title', 'match_score']])
    print("-" * 50)

# Fonksiyonu çalıştır
show_recommendation_logic()
```
